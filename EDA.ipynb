```python
import pandas as pd

# Load the dataset
file_path = 'C:/Users/任正/Documents/CAPSTONE/Capstone/ctg-studies - 400K.csv'
data = pd.read_csv(file_path)
```

```python
# Display basic information about the dataset
print(data.info())

# Display the first few rows of the dataset
print(data.head())


# Summary statistics for numerical columns
print(data.describe())

# Checking for missing values
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the distribution of 'Sex' column
plt.figure(figsize=(8, 6))
sns.countplot(data['Sex'])
plt.title('Distribution of Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

# Plotting the distribution of 'Study Status' column
plt.figure(figsize=(10, 6))
sns.countplot(data['Study Status'])
plt.title('Distribution of Study Status')
plt.xlabel('Study Status')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
